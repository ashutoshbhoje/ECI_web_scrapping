```python
import requests
from bs4 import BeautifulSoup

# Define the start URL
start_url = "https://affidavit.eci.gov.in/"

# Define a function to scrape data from a single page
candidate = []
state = []
party = []
AC_name = []
Status = []
def scrape_page(url):
    response = requests.get(url)
    if response.status_code == 200:
        soup = BeautifulSoup(response.content, "html.parser")

        # Scrape data from the current page
        h4_elements = soup.find_all("h4", class_=["bg-dark-blu", "bg-grn", "bg-yel", "bg-blu", "bg-red"])
        for h4_element in h4_elements:
            h4_text = h4_element.get_text()
            candidate.append(h4_text)
            print(h4_text)

    
            
        party_elements = [p for p in soup.find_all('p') if 'Party :' in p.text]
        party_names = [element.text.split('Party :')[1].strip() for element in party_elements]
        party.extend(party_names)
        print(party_names)

        state_elements = [p for p in soup.find_all('p') if 'State :' in p.text]
        state_names = [element.text.split('State :')[1].strip() for element in state_elements]
        state.extend(state_names)
        print(state_names)

        constituency_elements = [p for p in soup.find_all('p') if 'Constituency :' in p.text]
        constituency_names = [element.text.split('Constituency :')[1].strip() for element in constituency_elements]
        AC_name.extend(constituency_names)
        print(constituency_names)
        
        status_elements = [p for p in soup.find_all('p') if 'Status :' in p.text]
        status_info = [element.find('strong', text='Status :').find_next('strong').text for element in status_elements]
        for status in status_info:
            Status.append(status)
        print(status_info)

        # Find the link to the next page
#         next_page_link = soup.find("a", class_="page-link")
#         print(next_page_link)

        # Check if you've reached the maximum number of pages to scrape or if there is no next page
#         if max_pages >= 1 and next_page_link:
#             next_page_url = next_page_link.get("href")
#             print(next_page_url)
#             scrape_page(next_page_url, max_pages - 1)
#         else:
#             print("Scraping completed")

# Define the maximum number of pages to scrape



for x in range(1,1084):
    print("page - "+ str(x))
    
    url = "http://affidavit.eci.gov.in?page="+str(x)
    scrape_page(url)
  
```
